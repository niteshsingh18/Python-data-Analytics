```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
path = 'data/canada.xlsx'
df = pd.read_excel(path, skiprows=20, sheet_name=1, skipfooter=2)
df
```

```python
pd.set_option('display.max_columns', None) # optional
```

```python
# adding a column
```

```python
years = list(range(1980,2014))
print(years)
```

```python
df[years].sum(axis=1)
```

```python

df
```

```python
df['Total'] = df[years].sum(axis=1)
df
```

```python
print(df.columns.tolist())
```

```python
# type, coverage, AREA, REG, DEV
df.drop(columns=['Type', 'Coverage', 'AREA', 'REG', 'DEV'], inplace=True)
df
```

```python
# Renaming

df.rename(mapper={
    'OdName' : 'Country',
    'AreaName' : 'Continent',
    'RegName' : 'Region',
    'DevName' : 'Status'
}, axis=1, inplace=True)
df
```

```python
# All countries with 'A' with filter

df[df.Country.str.startswith("A")]
```

```python
# All countries ending with 'i' using query() 

df.query("Country.str.endswith('i')")
```

```python
# Countries with total imigration less than 10000

df.query("Total < 10000")
```

```python
# indexing 
df.set_index('Country', inplace=True)
```

```python
df.loc['Japan']
```

```python
df.loc['Japan', years].plot(figsize=(15,5), style='--o')
```

```python
df.loc['India', years].plot(figsize=(15,5), style='--o')
```

```python
df.loc['China', years].plot(figsize=(15,5), style='--o')
```

```python
df.loc[['India','China','Japan'], years].T.plot(figsize=(15,5), kind = 'area',
                                               stacked=False, alpha=.5)
```

```python
df.groupby('Continent')['Total'].sum()
```

```python
plt.figure(facecolor='White')
colors = ['#0f8','#f80', '#0f5', '#0f6', '#0f7', '#0d3']
df.groupby('Continent')['Total'].sum().plot(kind='pie', autopct='%.1f%%',
                                            colors = colors,
                                            figsize=(8,8),wedgeprops={
                                                'width' : .6,
                                                'edgecolor' : 'black'
                                            })

plt.title(f"'Half' of the immigrants to canada are from Asia", fontsize=20)
```

```python
sns.displot(data=df.query('Total>25000'), x = 'Total', kde=True, col='Status')
```
