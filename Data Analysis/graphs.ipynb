```python
rainfall = np.array([399, 555, 464, 587, 559])
years = np.array([2019,2020,2021,2022,2023])
# line plot
plt.plot(rainfall)
```

```python
plt.plot(years)
```

```python
# configuration graph
plt.plot(years,rainfall)
plt.title("Rainfall per year", fontdict={'fontsize':15,'color':'pink'})
plt.xlabel("years")
plt.ylabel("Rainfall (cm)")
plt.xticks(years)
plt.hlines(y=500, xmin=2019, xmax=2023, color='red', linestyles='--')
plt.show()
```

```python
plt.figure(figsize=(10,5), facecolor='lightblue')
x = np.array(['Rohan', 'Raju', 'Ram', 'Ranjan', 'Ramesh'])
y = np.array([60,70,80,90,100])
plt.bar(x,y)
plt.title("Salary Chart")
plt.show()
```

```python
plt.figure(figsize=(10,5), facecolor='lightblue')
x = np.array(['Rohan', 'Raju', 'Ram', 'Ranjan', 'Ramesh'])
y = np.array([60,70,80,90,100])
bars = plt.bar(x, y, 
        color = ['pink', 'red', 'blue', 'orange', 'purple'],
        alpha=.8, edgecolor='black', linewidth=2, hatch='/', width=.4)
bars[2].set_hatch('o')
bars[2].set_color('blue')
plt.title("Salary Chart")
plt.ylim(0,120)
for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2,
             bar.get_height()+10,
             f"${bar.get_height()}k",
             fontsize=10, color='red',
             ha = 'center')


plt.show()
```

```python
plt.pie(
    y,
    labels = x,
    explode = [0, 0, 0.1, 0, 0],
    autopct = '%.1f%%',
    pctdistance = .75,
    counterclock=True,
    startangle = 90,
    wedgeprops = {'width':.7},  # donut chart
    )
plt.show()
```
