```python
import seaborn as sns
import numpy as numpy
```

```python
x = np.random.normal()
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.set_theme(style='darkgrid')
```

```python
sns.relplot(data=tips, x='total_bill', y = 'tip', aspect = 1.5, height=4)
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue='sex')
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue='sex',
            palette=['y', 'g'], edgecolor='black', aspect=2)
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue='smoker', style='smoker',
            palette=['y', 'g'], edgecolor='black', aspect=2)
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue='time',
            palette=['y', 'g'], edgecolor='black', aspect=2)
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue='day',
            palette=['y', 'g', 'b', 'r'], edgecolor='black', aspect=2)
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", style="smoker"
)
```

```python
sns.relplot(
    data=tips, x="total_bill", y="tip", hue="size",
)
```

```python
sns.relplot(
    data=tips, x="total_bill", y="tip", hue="size",palette='hot'
)
```

```python
sns.relplot(
    data=tips, x='total_bill', y='tip',
    size='total_bill', aspect=3,
    sizes=(20,200), alpha=.5, edgecolor='blue'
)
```
